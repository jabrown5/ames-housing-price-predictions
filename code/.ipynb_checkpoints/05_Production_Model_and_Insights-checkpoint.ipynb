```python
# def metrics_ridge(dataset, target_var, features_list, test_dataset):
# metrics_ridge(train_data, 'SalePrice', features_X, test_data)

X = ames_train[features_X]
y = ames_train['SalePrice']

# instantiate the model
model_full = LinearRegression()

# 'Teaching' the model the features and y 
model_full.fit(X, y)

print(f"Baseline / Null Model: {y.mean()}")
print(f"Intercept: {model_full.intercept_}")   
print(f"R2: {model_full.score(X, y)}")   
print(f"R2adj: {1 - ((1 - (model_full.score(X, y)))*(len(y)-1) / (len(y) - len(features_X) - 1))}")
print(f'Ridge RMSE: {mean_squared_error(y, model_full.predict(X))**0.5}')

print('=============================')


X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    random_state=54, 
                                                    test_size=0.4,)

# Data Scaling
sc = StandardScaler()
Z = sc.fit_transform(X)

Z_train = sc.fit_transform(X_train)
Z_test = sc.transform(X_test)    

#test_dataset_sc = sc.transform(test_dataset)

# Instantiate Ridge
ridge = Ridge(alpha=10)

# Fit.
ridge.fit(Z_train, y_train)

y_preds = ridge.predict(Z)
residuals = y - y_preds

# Evaluate model using R2.
print(f'Ridge Baseline / Null Model: {y.mean()}')
print(f'Ridge Full Model Intercept: {ridge.intercept_}')  
print(f'Ridge Full Model R2 {ridge.score(Z, y)}')
print(f'Ridge Full Model R2adj: {1 - ((1 - (ridge.score(Z_train, y_train)))*(len(y)-1) / (len(y) - len(features_X) - 1))}')
print(f'Ridge Full Model RMSE: {mean_squared_error(y, ridge.predict(Z))**0.5}')

# Cross-validate over our list of ridge alphas.
ridge_cv = RidgeCV(alphas = (np.logspace(0, 3, 20)),
                  scoring='r2', # score we want
                  cv=10 # how many cross validations to run
                  )

# Fit model using best ridge alpha!
ridge_cv.fit(Z_train, y_train)
# Finding the best alpha      
ridge_cv.alpha_  
print(f"Ridge Train Baseline / Null Model: {y_train.mean()}")
print(f"Ridge Intercept: {ridge_cv.intercept_}")  
print(f'Ridge R2 Train: {ridge_cv.score(Z_train, y_train)}')
print(f'Ridge R2 Test: {ridge_cv.score(Z_test, y_test)}')      
print(f'Ridge Training RMSE: {mean_squared_error(y_train, ridge_cv.predict(Z_train))**0.5}')
print(f'Ridge Testing RMSE: {mean_squared_error(y_test, ridge_cv.predict(Z_test))**0.5}')
print('   ')

coef_df_ridge = pd.DataFrame({
'column' : X.columns,
'coef'   : ridge_cv.coef_})

print(coef_df_ridge)

print('=============================')
```
