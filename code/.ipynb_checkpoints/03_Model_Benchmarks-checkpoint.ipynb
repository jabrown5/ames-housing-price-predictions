```python
# Function to calculate R2-adjusted
# Used in following functions
def r2_adj(r2, y, features_list):
    r2_a = 1 - ((1 - r2)*(len(y)-1) / (len(y) - len(features_list) - 1))
    return r2_a
```

```python
# Function to run models: plain MLR w/ no transformations
# Source: Patrick's walk-through of Kaggle submission
def metrics(dataset, target_var, features_list, test_dataset):

    X = dataset[features_list]
    y = dataset[target_var]
    
    # instantiate the model
    model_full = LinearRegression()


    # Data Scaling
    sc = StandardScaler()
    Z = sc.fit_transform(X)

#     Z_train = sc.fit_transform(X_train)
#     Z_test = sc.transform(X_test)    

    test_dataset_sc = sc.transform(test_dataset)
    
    # 'Teaching' the model the features and y 
    model_full.fit(X, y)
    
    y_preds = model_full.predict(X)
    residuals = y - y_preds

    print(f"Baseline / Null Model: {y.mean()}")
    print(f"Intercept: {model_full.intercept_}")   
    print(f"R2: {model_full.score(X, y)}")   
    print(f"R2adj: {1 - ((1 - (model_full.score(X, y)))*(len(y)-1) / (len(y) - len(features_list) - 1))}")
    print(f'Ridge RMSE: {mean_squared_error(y, model_full.predict(X))**0.5}')
    print(f"Cross val R2: {cross_val_score(model_full, X, y).mean()}")
    print('    ')

    coef_df = pd.DataFrame({
    'column' : X.columns,
    'coef'   : model_full.coef_})
    
    print(coef_df)
    
    print('=============================')

    X_train, X_test, y_train, y_test = train_test_split(X, y,random_state=42)

    model = LinearRegression()
    model.fit(X_train, y_train)
    
    print(f"TRAIN Baseline / Null Model: {y_train.mean()}")
    print(f"Train Intercept: {model.intercept_}")  
    print(f"Training R2: {model.score(X_train, y_train)}")
    print(f"Testing R2: {model.score(X_test, y_test)}")
    print(f"Training RMSE: {mean_squared_error(y_train, model.predict(X_train))**0.5}")
    print(f"Testing RMSE: {mean_squared_error(y_test, model.predict(X_test))**0.5}")
    
    cross_val = cross_val_score(model, X_train, y_train).mean()
    
    # Plot the predicted y against the true y
    #return plt.scatter(y, y_preds)
    
    # Plot the residuals
    return plt.scatter(y_preds, residuals)

    # UNCOMMENT TO EXPORT CSV PREDICTION
    
#     model.fit(Z, y)
#     model_pred_final=model.predict(test_dataset_sc)
    
#     prediction = pd.DataFrame({'Id': test_id['Id'], 'SalePrice': model_pred_final}).to_csv('../datasets/test_prediction_last.csv', index=False)
```

```python
# Function to run models: MLR w/ RIDGE 
# Source: Patrick's walk-through of Kaggle submission
# Help from 'Intro to Regularization' lesson
def metrics_ridge(dataset, target_var, features_list, test_dataset):

    X = dataset[features_list]
    y = dataset[target_var]
    
    # instantiate the model
    model_full = LinearRegression()
    
    # 'Teaching' the model the features and y 
    model_full.fit(X, y)
    
    print(f"Baseline / Null Model: {y.mean()}")
    print(f"Intercept: {model_full.intercept_}")   
    print(f"R2: {model_full.score(X, y)}")   
    print(f"R2adj: {1 - ((1 - (model_full.score(X, y)))*(len(y)-1) / (len(y) - len(features_list) - 1))}")
    print(f'Ridge RMSE: {mean_squared_error(y, model_full.predict(X))**0.5}')

    print('=============================')


    X_train, X_test, y_train, y_test = train_test_split(X,
                                                        y,
                                                        random_state=54, 
                                                        test_size=0.4,)

    # Data Scaling
    sc = StandardScaler()
    Z = sc.fit_transform(X)

    Z_train = sc.fit_transform(X_train)
    Z_test = sc.transform(X_test)    

    test_dataset_sc = sc.transform(test_dataset)

    # Instantiate Ridge
    ridge = Ridge(alpha=10)

    # Fit.
    ridge.fit(Z_train, y_train)
    
    y_preds = ridge.predict(Z)
    residuals = y - y_preds
    
    # Evaluate model using R2.
    print(f'Ridge Baseline / Null Model: {y.mean()}')
    print(f'Ridge Full Model Intercept: {ridge.intercept_}')  
    print(f'Ridge Full Model R2 {ridge.score(Z, y)}')
    print(f'Ridge Full Model R2adj: {1 - ((1 - (ridge.score(Z_train, y_train)))*(len(y)-1) / (len(y) - len(features_list) - 1))}')
    print(f'Ridge Full Model RMSE: {mean_squared_error(y, ridge.predict(Z))**0.5}')

    # Cross-validate over our list of ridge alphas.
    ridge_cv = RidgeCV(alphas = (np.logspace(0, 3, 20)),
                      scoring='r2', # score we want
                      cv=10 # how many cross validations to run
                      )

    # Fit model using best ridge alpha!
    ridge_cv.fit(Z_train, y_train)
    # Finding the best alpha      
    ridge_cv.alpha_  
    print(f"Ridge Train Baseline / Null Model: {y_train.mean()}")
    print(f"Ridge Intercept: {ridge_cv.intercept_}")  
    print(f'Ridge R2 Train: {ridge_cv.score(Z_train, y_train)}')
    print(f'Ridge R2 Test: {ridge_cv.score(Z_test, y_test)}')      
    print(f'Ridge Training RMSE: {mean_squared_error(y_train, ridge_cv.predict(Z_train))**0.5}')
    print(f'Ridge Testing RMSE: {mean_squared_error(y_test, ridge_cv.predict(Z_test))**0.5}')
    print('   ')

    coef_df_ridge = pd.DataFrame({
    'column' : X.columns,
    'coef'   : ridge_cv.coef_})

    print(coef_df_ridge)
    
    print('=============================')
   


    # UNCOMMENT THE BELOW TO PRODUCE EXPORT CSV FOR KAGGLE
    ridge_cv.fit(Z, y)
    model_pred_final=ridge_cv.predict(test_dataset_sc)
    
    prediction = pd.DataFrame({'Id': test_id['Id'], 'SalePrice': model_pred_final}).to_csv('../datasets/test_prediction_last.csv', index=False)

    # Plot the predicted y against the true y
    return plt.scatter(y, y_preds)

    # Plot the residuals
    #return plt.scatter(y_preds, residuals)
```

```python
# train_data = ames_train[features]
# test_data = ames_test[features_X]

metrics_ridge(train_data, 'SalePrice', features_X, test_data)
```

```python
# Function to run models: MLR w/ LASSO 
# Source: Patrick's walk-through of Kaggle submission
# Help from 'Intro to Regularization' lesson
def metrics_lasso(dataset, target_var, features_list, test_dataset):

    X = dataset[features_list]
    y = dataset[target_var]

    
    # instantiate the model
    model_full = LinearRegression()

    
    # 'Teaching' the model the features and y 
    model_full.fit(X, y)
    
    print(f"Baseline / Null Model: {y.mean()}")
    print(f"Intercept: {model_full.intercept_}")   
    print(f"R2: {model_full.score(X, y)}")   
    print(f"R2adj: {1 - ((1 - (model_full.score(X, y)))*(len(y)-1) / (len(y) - len(features_list) - 1))}")
    print(f'Ridge RMSE: {mean_squared_error(y, model_full.predict(X))**0.5}')
    print('=============================')

    X_train, X_test, y_train, y_test = train_test_split(X, 
                                                        y,
                                                        random_state=54, 
                                                        test_size=0.4,)

    # Data Scaling
    sc = StandardScaler()
    Z = sc.fit_transform(X)

    Z_train = sc.fit_transform(X_train)
    Z_test = sc.transform(X_test)    

    test_dataset_sc = sc.transform(test_dataset)
    
    y_preds = model_full.predict(Z)
    residuals = y - y_preds

    
    print('=============================')
   
    # LASSO

    # Cross-validate over our list of Lasso alphas.
    # doing 5 fold cross validation
    lasso_cv = LassoCV(alphas=(np.logspace(-3, 0, 100)), cv=5)

    # Fit model using best CV
    lasso_cv.fit(Z_train, y_train)

    # Here is the optimal value of alpha
    lasso_cv.alpha_
    
    print(f"Lasso Train Baseline / Null Model: {y_train.mean()}")
    print(f"Lasso Intercept: {lasso_cv.intercept_}")  
    print(f'Lasso R2 Full: {lasso_cv.score(Z, y)}')
    print(f'Lasso Full RMSE: {mean_squared_error(y, lasso_cv.predict(Z))**0.5}')
    print(f'Lasso R2 Train: {lasso_cv.score(Z_train, y_train)}')
    print(f'Lasso R2 Test: {lasso_cv.score(Z_test, y_test)}')
    print(f'Lasso Training RMSE: {mean_squared_error(y_train, lasso_cv.predict(Z_train))**0.5}')
    print(f'Lasso Testing RMSE: {mean_squared_error(y_test, lasso_cv.predict(Z_test))**0.5}')
    print('   ')
    
    # Finding the Lasso coefficients
    coef_df_lasso = pd.DataFrame({
    'column' : X.columns,
    'coef'   : lasso_cv.coef_})
    
    print(coef_df_lasso)
    


    # UNCOMMENT THE BELOW TO PRODUCE EXPORT CSV FOR KAGGLE
    lasso_cv.fit(Z, y)
    model_pred_final=lasso_cv.predict(test_dataset_sc)
    
    prediction = pd.DataFrame({'Id': test_id['Id'], 'SalePrice': model_pred_final}).to_csv('../datasets/test_prediction_last.csv', index=False)

    # Plot the residuals
    return plt.scatter(y_preds, residuals)

    # Plot the predicted y against the true y
    #return plt.scatter(y, y_preds)
    
```

```python
# Calling the Lasso Function
metrics_lasso(train_data, 'SalePrice', features_X, test_data)
```

```python
poly = PolynomialFeatures(degree=2, interaction_only=False, include_bias=False)

# Function to run models - MLR w/ POLY and LASSO
# Source: Patrick's walk-through of Kaggle submission
# Help from 'Intro to Regularization' lesson
# THIS MODEL WAS DESIGNED TO EXAMINE POLYNOMIAL FEATURES AND WHETHER TO INCLUDE THEM
def metrics_poly(dataset, target_var, features_list, test_dataset):

    X = dataset[features_list]
    y = dataset[target_var]

    # create only 2x interactions
    poly = PolynomialFeatures(degree=2, interaction_only=False, include_bias=False)
   
    # Fit / transform
    X_overfit = poly.fit_transform(X)
    
    X_train, X_test, y_train, y_test = train_test_split(X_overfit, 
                                                        y, 
                                                        random_state=8)    
    # Standard Scaler
    sc = StandardScaler()
    Z_train = sc.fit_transform(X_train)
    Z_test = sc.transform(X_test)
    Z = sc.fit_transform(X_overfit)
    
    test_dataset_sc = sc.transform(test_dataset)
   
    lr = LinearRegression()
    lr.fit(Z_train, y_train)
        
    y_preds = lr.predict(Z)
    residuals = y - y_preds
    
    print(f"Baseline / Null Model: {y_train.mean()}")
    print(f"Intercept: {lr.intercept_}")   
    print(f'Poly Train R2: {lr.score(Z_train, y_train)}')
    print(f'Poly Test R2: {lr.score(Z_test, y_test)}')
    print(f'Poly Train RMSE: {mean_squared_error(y_train, lr.predict(Z_train))**0.5}')
    print(f'Poly Test RMSE: {mean_squared_error(y_test, lr.predict(Z_test))**0.5}')    
    print('============================')
          
   
    # LASSO
    
    # Cross-validate over our list of Lasso alphas.
    # doing 5 fold cross validation
    lasso_cv = LassoCV(alphas=(np.logspace(-3, 0, 100)), cv=5)

    # Fit model using best CV
    lasso_cv.fit(Z_train, y_train)

    # Here is the optimal value of alpha
    print(f"Lasso's Best Alpha: {lasso_cv.alpha_}")
    print('=========================')
    print(f'Lasso R2 Train: {lasso_cv.score(Z_train, y_train)}')
    print(f'Lasso R2 Test: {lasso_cv.score(Z_test, y_test)}')
    print(f'Lasso Training RMSE: {mean_squared_error(y_train, lasso_cv.predict(Z_train))**0.5}')
    print(f'Lasso Testing RMSE: {mean_squared_error(y_test, lasso_cv.predict(Z_test))**0.5}')
    print('=========================')
    print(f'Lasso R2 Full: {lasso_cv.score(Z, y)}')
    print(f'Lasso Full RMSE: {mean_squared_error(y, lasso_cv.predict(Z))**0.5}')

    # Getting feature names:
    # https://stackoverflow.com/questions/36728287/sklearn-preprocessing-polynomialfeatures-how-to-keep-column-names-headers-of
    coef_df_lasso = pd.DataFrame({
    'column' : poly.get_feature_names(X.columns),
    'coef'   : lasso_cv.coef_})
    
    print(coef_df_lasso)

    lasso_cv.fit(X_overfit, y)
    model_pred_final=lasso_cv.predict(test_dataset_sc)
    
    prediction = pd.DataFrame({'Id': test_id['Id'], 'SalePrice': model_pred_final}).to_csv('../datasets/test_prediction_last.csv', index=False)

    # Prints the y true against y predictions
    #return plt.scatter(y, y_preds)
    
    # Prints residual spread
    return plt.scatter(y_preds, residuals)
```

```python
# Calling the metrics function that tranforms variables through interactions (Polynomial)
metrics_poly(train_data, 'SalePrice', features_X, poly.fit_transform(test_data))
```

```python
# Function to run models:  MLR w/ KNN
# Source: Patrick's walk-through of Kaggle submission
# Help from 'Intro to Classification and KNN' and 'Hyperparameters, Gridsearch, and Pipelines' lessons
def metrics_knn(dataset, target_var, features_list, test_dataset):

    X = dataset[features_list]
    y = dataset[target_var]
    
    # instantiate the model
    model_full = LinearRegression()
    
    # 'Teaching' the model the features and y 
    model_full.fit(X, y)
    
    y_preds = model_full.predict(X)
    residuals = y - y_preds
    
    print(f"Intercept: {model_full.intercept_}")   
    print(f"R2: {model_full.score(X, y)}")   
    print(f"R2adj: {1 - ((1 - (model_full.score(X, y)))*(len(y)-1) / (len(y) - len(features_list) - 1))}")
    print('============================')
    
    X_train, X_test, y_train, y_test = train_test_split(X, 
                                                        y,
                                                        random_state=42)

    
    # Data Scaling
    sc = StandardScaler()
    Z = sc.fit_transform(X)
    Z_train = sc.fit_transform(X_train)
    Z_test = sc.transform(X_test)    

    # Setting up KNN parameters
    knn_params = {
        'n_neighbors': range(1, 51, 2),
        'metric': ['euclidean', 'manhattan']
    }    

    # Gridsearch using paramters set up above
    knn_gridsearch = GridSearchCV(estimator = KNeighborsRegressor(), 
                              param_grid = knn_params, 
                              cv = 5, 
                              verbose=1)
    
    knn_gridsearch.fit(Z_train, y_train)
    
    #print(f'KNN best R2: {knn_gridsearch.best_score_}')
    print(f'KNN best parameters: {knn_gridsearch.best_params_}')
    print(f'KNN R2 Train: {knn_gridsearch.score(Z_train, y_train)}')
    print(f'KNN R2 Test: {knn_gridsearch.score(Z_test, y_test)}')

    print(f'KNN Training RMSE: {mean_squared_error(y_train, knn_gridsearch.predict(Z_train))**0.5}')
    print(f'KNN Testing RMSE: {mean_squared_error(y_test, knn_gridsearch.predict(Z_test))**0.5}')
    print('========================')
    
    knn = KNeighborsRegressor(metric='manhattan', n_neighbors=7)
    knn.fit(Z_train,y_train)

    print(f"Baseline / Null Model: {Z.mean()}")
     
    print(f'KNN final R2 Train: {knn.score(Z_train, y_train)}')
    print(f'KNN final R2 Test: {knn.score(Z_test, y_test)}')

    print(f'KNN final Training RMSE: {mean_squared_error(y_train, knn.predict(Z_train))**0.5}')
    print(f'KNN final Testing RMSE: {mean_squared_error(y_test, knn.predict(Z_test))**0.5}')
    
    # Print the y true against y predicted values
    #return plt.scatter(y, y_preds)
    
    # Print the residuals
    return plt.scatter(y_preds, residuals)

```

```python
# Calling the KNN function
metrics_knn(train_data, 'SalePrice', features_X, test_data)
```
