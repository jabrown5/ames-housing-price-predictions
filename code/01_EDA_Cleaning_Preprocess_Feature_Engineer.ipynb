```python
# Creating a function to apply same changes to TRAIN and TEST datasets
def scale_columns(dataset):
    # Mapping some ordinal values to numeric instead of text
    
    # Exterior Quality - making ordinal
    dataset['Exter Qual'] = dataset['Exter Qual'].map({'Po':1,'Fa':2,'TA':3,'Gd':4,'Ex':5})
    # Exterior Condition - making ordinal   
    dataset['Exter Cond'] = dataset['Exter Cond'].map({'Po':1,'Fa':2,'TA':3,'Gd':4,'Ex':5})
    # Basement quality - filling missing values with NA
    dataset['Bsmt Qual'] = dataset['Bsmt Qual'].fillna(value='NA')
    # Basement quality - making ordinal
    dataset['Bsmt Qual ord'] = dataset['Bsmt Qual'].map({'NA':0,'Po':1,'Fa':2,'TA':3,'Gd':4,'Ex':5})
    # Basement Quality - filling any missing values w/ 0 to indicate no basement
    dataset['Bsmt Qual ord'] = dataset['Bsmt Qual ord'].fillna(value=0)

    
    dataset['Bsmt Cond'] = dataset['Bsmt Cond'].fillna(value='NA')    
    dataset['Bsmt Cond ord'] = dataset['Bsmt Cond'].map({'NA':0,'Po':1,'Fa':2,'TA':3,'Gd':4,'Ex':5})
    dataset['Bsmt Cond ord'] = dataset['Bsmt Cond ord'].fillna(value=0)

    dataset['Bsmt Qual'] = dataset['Bsmt Qual'].map({'NA':'NA','Po':'Poor','Fa':'Fair','TA':'Typical','Gd':'Good','Ex':'Excellent'})
    dataset['Bsmt Cond'] = dataset['Bsmt Cond'].map({'NA':'NA','Po':'Poor','Fa':'Fair','TA':'Typical','Gd':'Good','Ex':'Excellent'})

    dataset['BsmtFin Type 1'] = dataset['BsmtFin Type 1'].fillna(value='NA')    
    dataset['BsmtFin Type 1'] = dataset['BsmtFin Type 1'].map({'NA':'NA','Unf':'Unfinished','LwQ':'Low_Qual','Rec':'Avg_Rec_Room','BLQ':'Below_Avg_Liv_Q','ALQ':'Avg_Liv_Q','GLQ':'Good_Liv_Q'})
    
    dataset['BsmtFin Type 2'] = dataset['BsmtFin Type 2'].fillna(value='NA')    
    dataset['BsmtFin Type 2'] = dataset['BsmtFin Type 2'].map({'NA':'NA','Unf':'Unfinished','LwQ':'Low_Qual','Rec':'Avg_Rec_Room','BLQ':'Below_Avg_Liv_Q','ALQ':'Avg_Liv_Q','GLQ':'Good_Liv_Q'})

    dataset['Bsmt Exposure'] = dataset['Bsmt Exposure'].map({'NA':'NA','No':'None','Mn':'Min','Av':'Avg','Gd':'Good'})
    dataset['Bsmt Exposure'] = dataset['Bsmt Exposure'].fillna(value='NA')

    dataset['Heating QC'] = dataset['Heating QC'].map({'Po':1,'Fa':2,'TA':3,'Gd':4,'Ex':5})
    dataset['Central Air'] = dataset['Central Air'].map({'N':0,'Y':1})

    dataset['Kitchen Qual'] = dataset['Kitchen Qual'].map({'Po':1,'Fa':2,'TA':3,'Gd':4,'Ex':5})
    
    dataset['Garage Qual ord'] = dataset['Garage Qual'].map({'NA':0,'Po':1,'Fa':2,'TA':3,'Gd':4,'Ex':5})
    dataset['Garage Qual ord'] = dataset['Garage Qual ord'].fillna(value=0)
    
    dataset['Garage Cond ord'] = dataset['Garage Cond'].map({'NA':0,'Po':1,'Fa':2,'TA':3,'Gd':4,'Ex':5})
    dataset['Garage Cond ord'] = dataset['Garage Cond ord'].fillna(value=0)
    
    dataset['Garage Qual'] = dataset['Garage Qual'].map({'NA':'NA','Po':'Poor','Fa':'Fair','TA':'Typical','Gd':'Good','Ex':'Excellent'})
    dataset['Garage Qual'] = dataset['Garage Qual'].fillna(value='NA')

    dataset['Garage Cond'] = dataset['Garage Cond'].map({'NA':'NA','Po':'Poor','Fa':'Fair','TA':'Typical','Gd':'Good','Ex':'Excellent'})
    dataset['Garage Cond'] = dataset['Garage Cond'].fillna(value='NA')

    dataset['Garage Type'] = dataset['Garage Type'].map({'NA':'None','Detchd':'Detached','CarPort':'CarPort','BuiltIn':'BuiltIn','Basment':'Basment','Attchd':'Attchd','2Types':'2Types'})
    dataset['Garage Type'] = dataset['Garage Type'].fillna(value='NA')

    dataset['Garage Finish'] = dataset['Garage Finish'].map({'nan':'None','Unf':'Unf','RFn':'RoughFin','Fin':'Fin'})
    dataset['Garage Finish'] = dataset['Garage Finish'].fillna(value='NA')


    dataset['Lot Shape'] = dataset['Lot Shape'].map({'Reg':'Reg','IR1':'Slight_Irreg','IR2':'Mod_Irreg','IR3':'Irreg'})
    dataset['Land Contour'] = dataset['Land Contour'].map({'Lvl':'Lvl','Bnk':'Banked','HLS':'Hillside','Low':'Depression'})

    dataset['Lot Config'] = dataset['Lot Config'].map({'Inside':'Inside','Corner':'Corner','CulDSac':'CulDSac','FR2':'Front_2','FR3':'Front_3'})
    
    dataset['Functional'] = dataset['Functional'].map({'Sal':0,'Sev':1,'Maj2':2,'Maj1':3, 'Mod':4, 'Min2':5, 'Min1':6, 'Typ':7})

    # Creating Age variables
    dataset['Age'] = dataset['Yr Sold'] - dataset['Year Built']
    dataset['Age_Remod'] = dataset['Yr Sold'] - dataset['Year Remod/Add']
    dataset['Remod_Yes'] = (dataset['Age_Remod'] > 0)*1

    # creating a garage dummy 1=yes, has a garage
    dataset['garage_yes'] = [0 if garage == 0 else 1 for garage in dataset['Garage Qual']]
    # creating a basement dummy 1=yes, has a garage
    dataset['basement_yes'] = [0 if garage == 0 else 1 for garage in dataset['Bsmt Qual']]

    # Preparing Paved Drive for dummification
    dataset['Paved Drive'] = dataset['Paved Drive'].map({'Y':'Paved','P':'Paved_Partial','N':'Dirt_Gravel'})

    # Creating Binary / Dummy variable to indicate whether a home has one of the below porch features
    dataset['Wood_Deck_Yes'] = (dataset['Wood Deck SF'] > 0)*1
    dataset['Open_Porch_Yes'] = (dataset['Open Porch SF'] > 0)*1
    dataset['Enclosed_Porch_Yes'] = (dataset['Enclosed Porch'] > 0)*1
    dataset['3Ssn_Porch_Yes'] = (dataset['3Ssn Porch'] > 0)*1
    dataset['Screen_Porch_Yes'] = (dataset['Screen Porch'] > 0)*1
    # Dummy variable to flag if a home is a new build
    dataset['Home_New'] = (dataset['Year Built'] == dataset['Yr Sold'] )*1

    # fill nan's w/ 0s for garage cars
    dataset['Garage Cars'] = dataset['Garage Cars'].fillna(value=0)
    # fill nan's w/ 0s for garage area 
    dataset['Garage Area'] = dataset['Garage Area'].fillna(value=0)
    # filling missing lot frontage values w/ 0
    dataset['Lot Frontage'] = dataset['Lot Frontage'].fillna(value=0)
    # filling basement variables that are missing w/ 0 to indicate none of the each
    dataset[['Bsmt Full Bath','Bsmt Half Bath','BsmtFin SF 1']] = dataset[['Bsmt Full Bath','Bsmt Half Bath','BsmtFin SF 1']].fillna(value=0)
    dataset[['Bsmt Unf SF','Total Bsmt SF','BsmtFin SF 2']] = dataset[['Bsmt Unf SF','Total Bsmt SF','BsmtFin SF 2']].fillna(value=0)
  
```

```python
# Setting up a function to create interactions indicated by Polynomial Regression as having importance
def create_interactions(dataset):
    dataset['Yr Sold X Exter Qual'] = dataset['Yr Sold'] * dataset['Exter Qual']
    dataset['Yr Sold X Age'] = dataset['Yr Sold'] * dataset['Age']
    dataset['Exter Qual X Age'] = dataset['Exter Qual'] * dataset['Age']
    dataset['Yr Sold X Bedroom AbvGr'] = dataset['Yr Sold'] * dataset['Bedroom AbvGr']
    dataset['Exter Cond X Lot Area'] = dataset['Exter Cond'] * dataset['Lot Area']
    dataset['Yr Sold X Lot Area'] = dataset['Yr Sold'] * dataset['Lot Area']
    dataset['Exter Cond X Kitchen Qual'] = dataset['Exter Cond'] * dataset['Kitchen Qual']
    dataset['Home_New X BsmtFin SF 1'] = dataset['Home_New'] * dataset['BsmtFin SF 1']
    dataset['Exter Qual X Overall Cond'] = dataset['Exter Qual'] * dataset['Overall Cond']
    dataset['Kitchen Qual X Age'] = dataset['Kitchen Qual'] * dataset['Age']
    dataset['Kitchen AbvGr X Lot Area'] = dataset['Kitchen AbvGr'] * dataset['Lot Area']
    dataset['Kitchen Qual X Kitchen AbvGr'] = dataset['Kitchen Qual'] * dataset['Kitchen AbvGr']
    dataset['Exter Cond X Overall Cond'] = dataset['Exter Cond'] * dataset['Overall Cond']
    dataset['Exter Qual X Total Bsmt SF'] = dataset['Exter Qual'] * dataset['Total Bsmt SF']
    dataset['Kitchen AbvGr X Overall Cond'] = dataset['Kitchen AbvGr'] * dataset['Overall Cond']
    dataset['Exter Cond X Gr Liv Area'] = dataset['Exter Cond'] * dataset['Gr Liv Area']
    dataset['Home_New X Full Bath'] = dataset['Home_New'] * dataset['Full Bath']
    dataset['Exter Qual X Heating QC'] = dataset['Exter Qual'] * dataset['Heating QC']
    dataset['Exter Qual X Gr Liv Area'] = dataset['Exter Qual'] * dataset['Gr Liv Area']
    dataset['Home_New X Bsmt Full Bath'] = dataset['Home_New'] * dataset['Bsmt Full Bath']
    dataset['Yr Sold X Gr Liv Area'] = dataset['Yr Sold'] * dataset['Gr Liv Area']

# Applying mapping function to ames_train dataset
create_interactions(ames_train)

# Applying mapping function to ames_test dataset
create_interactions(ames_test)
```
